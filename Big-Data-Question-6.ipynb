```python
# Import Libraries
import pandas as pd
import boto3


# Defining variables that will be needed to read the csv file
bucket = "tolubucket1"
file_name = "filtered_data.csv"


# Creating the low level functional client (since I'm only using 1 bucket)
s3 = boto3.client('s3') 

# Create the S3 object
obj = s3.get_object(Bucket= bucket, Key= file_name) 

# Read data from the S3 object
data = pd.read_csv(obj['Body'])

print(data)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
plt.plot(data['year'], data['frequency'])
plt.title("Number of Occurrences of the Token Over the Years")
plt.xlabel("Year")
plt.ylabel("Occurrences of the Token(Frequency)")
plt.show()
```
